```python
import pandas as pd
```

# Boston Data

```python
# 1
# Read the data into a dataframe 
boston_data = pd.read_csv("../data/mbta_untidy.csv")
boston_data
```

```python
# 2 
# Drop 'order' column from dataframe
boston_data.drop(columns=["order"], inplace=True)
boston_data
```

```python
#3 
# Use the melt function to convert the “year and month” columns to a single column called
# “year_mon” in the dataframe.

boston_data_long = pd.melt(boston_data, id_vars=["mode"], var_name="year_mon",
                          value_name="riders")
boston_data_long
```

```python
# 4
# Split the values in the “year_mon” column into year and month values, then add those values to the
# dataframe as appropriately named columns.

variable_split = boston_data_long["year_mon"].str.split("-")
variable_split.head()
```

```python
# Extract the values needed

year_values = variable_split.str.get(0)
month_values = variable_split.str.get(1)

print(year_values.describe())
print(month_values.describe())
```

```python
# Now add them to the dataframe.
boston_data_long["year"] = year_values
boston_data_long["month"] = month_values

boston_data_long.head()
```

```python
# 5 
# Drop the “year_mon” column.

boston_data_long.drop(columns=["year_mon"], inplace=True) # inplace=True makes the change permanemt

boston_data_long
```

# London Heathrow Data

```python
# 1
# Read the data into a dataframe

london_data = pd.read_csv("../data/lhr_visit_len.csv")
london_data
```

```python
# 2 
# Melt the data columns into a column called “visit_length” and a column called “count.”

london_data_long = pd.melt(london_data, id_vars=["type"], var_name="visit_length",
                          value_name="count")
london_data_long
```

```python
# 3 
# Split the values in the “type” column into values for variables called “residence”, “purpose”, and
# “flight_type”.

variable_split = london_data_long["type"].str.split("_")
variable_split.head()
```

```python
# 3 continued 
# Extract the values needed

residence_values = variable_split.str.get(0)
purpose_values = variable_split.str.get(1)
flight_type_values = variable_split.str.get(2)

print(residence_values.describe())
print(purpose_values.describe())
print(flight_type_values.describe())
```

```python
# 3 continued
# Now add them to the dataframe.

london_data_long["residence"] = residence_values
london_data_long["purpose"] = purpose_values
london_data_long["flight_type"] = flight_type_values

london_data_long.head()
```

```python
# 4
# Drop the “type” column.

london_data_long.drop(columns=["type"], inplace=True) # inplace=True makes the change permanemt

london_data_long
```
