```python
import seaborn as sns
import pandas as pd
applewood_data = pd.read_csv("../data/Applewood_2011.csv")
applewood_data
```

```python
# 1
ax = sns.pairplot(applewood_data)
```

```python
# 2 
ax = sns.countplot(x="Location", data=applewood_data)
```

```python
# 3
ax = sns.boxplot(x="Previous", y="Profit", data=applewood_data)
```

```python
# 4
ax = sns.lmplot(x="Age",
                y="Profit",
                fit_reg=False,
                ci=False,
                col="Previous",
                col_wrap=3,
                data=applewood_data)
```

```python
# Dataset number 2
college = pd.read_csv("../data/College.csv")
# 1
ax = sns.lmplot(x="Outstate",
               y="Room.Board",
               col="Private",
               data=college)
```

```python
boston_housing = pd.read_csv("../data/Boston_housing.csv", sep=";")
```

```python
# Dataset number 3
# 1 
ax = sns.lmplot(x="LSTAT",
                y="MEDV",
                col="CHAS",
                ci=False,
                fit_reg=False,
                data=boston_housing)
```

```python
# 2
ax = sns.boxplot(x="CHAS",
                 y="MEDV",
                 data=boston_housing)
ax.set_title("Median Value by Charles River Proximity")
ax.set_xlabel("Near the Charles River?")
ax.set_ylabel("Median Home Value")
```

```python
# 3
ax = sns.scatterplot(x="RM",
                y="MEDV",
                data=boston_housing)
ax.set_title("Number of Rooms versus Median Value")
ax.set_xlabel("Number of Rooms")
ax.set_ylabel("Median House Value (in thousands)")
```
