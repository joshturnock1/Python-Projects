```python
taxes = pd.read_csv("../data/15zpnyagi.csv").squeeze("columns")
type(taxes)
```

```python
# getting data of all ny tax filers with income range of $50,000 to $70,000
taxes1 = ny_taxes[ny_taxes["agi_stub"] == 3]
# from this subset, now subsetting number of dependents
taxes_num_dep = taxes1["NUMDEP"]
taxes_num_dep
total_dependents = sum(taxes_num_dep)
total_dependents

# getting total dependents claimed from the subset of income range of $50,0000 - $70,000
total_returns = taxes1.index
sum(total_returns)

# dividing total dependents by total returns to get the mean
final = total_dependents / sum(total_returns)

# meaningful print statement
print(f"The average number of dependents per return in the specified AGI range is {final:.2f}")
```
