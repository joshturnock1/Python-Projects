```python
# import the libraries
import pandas as pd
import datetime as dt
import seaborn as sns
```

# 1

```python
# load the dataset with date column the index and parse dates

ibm = pd.read_csv("../data/IBM.csv", 
                 index_col = "Date", 
                  parse_dates = ["Date"])
ibm
```

# 2

```python
ibm_qtr = ibm.resample("Q").mean()
ibm_qtr
```

# 3

```python
# Using seaborn to create graph plotting closing price vs quarter designation

graph = sns.lineplot(x="Date",
                    y="Close",
                    data=ibm_qtr)
```

```python
# alternative way using plot() method
# creating graph plotting closing price vs quarter designation using plot() method
ibm_qtr["Close"].plot(y="Close")
```
